```python
import random
import os

def get_best_score():
    if os.path.exists("best_score.txt"):
        with open("best_score.txt", "r") as file:
            return int(file.read())
    return None

def save_best_score(score):
    with open("best_score.txt", "w") as file:
        file.write(str(score))

def play_game():
    print("\n===== NUMBER GUESSING GAME =====")

    print("Select Difficulty Level:")
    print("1. Easy (1-50, 10 attempts)")
    print("2. Medium (1-100, 7 attempts)")
    print("3. Hard (1-200, 5 attempts)")

    level = input("Enter choice (1/2/3): ")

    if level == '1':
        number = random.randint(1, 50)
        attempts = 10
    elif level == '2':
        number = random.randint(1, 100)
        attempts = 7
    elif level == '3':
        number = random.randint(1, 200)
        attempts = 5
    else:
        print("Invalid choice!")
        return

    best_score = get_best_score()
    if best_score:
        print("üèÜ Best Score (Least Attempts):", best_score)

    count = 0

    while count < attempts:
        try:
            guess = int(input(f"Attempt {count+1}/{attempts} - Enter your guess: "))
            count += 1

            if guess == number:
                print("üéâ Congratulations! You guessed correctly!")
                print("Attempts taken:", count)

                if best_score is None or count < best_score:
                    save_best_score(count)
                    print("üèÜ New Best Score Saved!")

                break
            elif guess < number:
                print("Too Low ‚¨á")
            else:
                print("Too High ‚¨Ü")

        except ValueError:
            print("Enter valid number!")

    else:
        print("‚ùå Game Over! The number was:", number)

while True:
    play_game()
    again = input("\nDo you want to play again? (yes/no): ").lower()
    if again != "yes":
        print("Thanks for playing üòä")
        break
```
