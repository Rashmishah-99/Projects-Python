```python
import tkinter as tk
from tkinter import messagebox
import json
import os

FILE_NAME = "users.json"

# ---------------- LOAD USERS ----------------
def load_users():
    if os.path.exists(FILE_NAME):
        with open(FILE_NAME, "r") as f:
            return json.load(f)
    return {}

# ---------------- SAVE USERS ----------------
def save_users(users):
    with open(FILE_NAME, "w") as f:
        json.dump(users, f, indent=4)

users = load_users()

# ---------------- REGISTER FUNCTION ----------------
def register():
    username = reg_username.get()
    password = reg_password.get()

    if not username or not password:
        messagebox.showerror("Error", "All fields required!")
        return

    if username in users:
        messagebox.showerror("Error", "Username already exists!")
        return

    users[username] = password
    save_users(users)

    messagebox.showinfo("Success", "Registration Successful!")
    reg_username.delete(0, tk.END)
    reg_password.delete(0, tk.END)

# ---------------- LOGIN FUNCTION ----------------
def login():
    username = login_username.get()
    password = login_password.get()

    if username in users and users[username] == password:
        messagebox.showinfo("Success", f"Welcome {username} ðŸŽ‰")
        open_dashboard(username)
    else:
        messagebox.showerror("Error", "Invalid Credentials!")

# ---------------- DASHBOARD ----------------
def open_dashboard(username):
    dashboard = tk.Toplevel(root)
    dashboard.title("Dashboard")
    dashboard.geometry("300x200")

    tk.Label(dashboard, text=f"Welcome {username}!",
             font=("Arial",14)).pack(pady=20)

    tk.Button(dashboard, text="Logout",
              command=dashboard.destroy).pack()

# ---------------- GUI SETUP ----------------
root = tk.Tk()
root.title("Login & Register System")
root.geometry("400x500")
root.configure(bg="#1e1e1e")

tk.Label(root, text="User Authentication System",
         font=("Arial",16,"bold"),
         bg="#1e1e1e", fg="cyan").pack(pady=10)

# ----- Register Section -----
tk.Label(root, text="Register",
         bg="#1e1e1e", fg="white").pack(pady=5)

reg_username = tk.Entry(root)
reg_username.pack(pady=5)
reg_username.insert(0, "Username")

reg_password = tk.Entry(root, show="*")
reg_password.pack(pady=5)

tk.Button(root, text="Register",
          command=register,
          bg="green", fg="white").pack(pady=5)

# ----- Login Section -----
tk.Label(root, text="Login",
         bg="#1e1e1e", fg="white").pack(pady=10)

login_username = tk.Entry(root)
login_username.pack(pady=5)

login_password = tk.Entry(root, show="*")
login_password.pack(pady=5)

tk.Button(root, text="Login",
          command=login,
          bg="blue", fg="white").pack(pady=5)

root.mainloop()
```
